```python
from transformers import AutoTokenizer
import matplotlib.pyplot as plt
import numpy as np
import ast
import pandas as pd
import os

dir = "masking_output"
# file = "entry_76_tokens_4512_response_2025-04-21_00-31-27.csv"
file = "entry_10_tokens_3490_response_2025-04-21_00-22-46.csv"
path = os.path.join(dir, file)
df = pd.read_csv(path)
df[df["is_correct"] == False]["attention_mask"].iloc[0]
num_values = df[df["is_correct"] == False]["attention_mask"].shape[0]
model_id = "deepseek-ai/DeepSeek-R1-Distill-Llama-8B"
tokenizer = AutoTokenizer.from_pretrained(model_id)
for value in range(num_values):
    # Get the data from the first incorrect answer row
    row_index = df[df["is_correct"] == False].index[value]
    attention_mask_str = df.loc[row_index, "attention_mask"]
    input_ids_str = df.loc[row_index, "input_ids"]

    # Convert from string representation to lists
    attention_mask_list = ast.literal_eval(attention_mask_str)
    input_ids_list = ast.literal_eval(input_ids_str)

    # Convert to numpy arrays
    mask_array = np.array(attention_mask_list[0])
    input_ids_array = np.array(input_ids_list[0])

    # Ensure both arrays have the same length
    min_length = min(len(mask_array), len(input_ids_array))
    mask_array = mask_array[:min_length]
    input_ids_array = input_ids_array[:min_length]

    # Calculate grid dimensions for a square-ish layout
    grid_size = int(np.sqrt(len(mask_array)))
    if grid_size**2 < len(mask_array):
        grid_size += 1

    # Create square grids for both arrays
    mask_grid = np.ones((grid_size**2), dtype=int) * -1  # -1 as padding value
    mask_grid[:len(mask_array)] = mask_array
    mask_grid = mask_grid.reshape(grid_size, grid_size)

    ids_grid = np.ones((grid_size**2), dtype=int) * -1  # -1 as padding value
    ids_grid[:len(input_ids_array)] = input_ids_array
    ids_grid = ids_grid.reshape(grid_size, grid_size)

    # Create the heatmap with input_ids as text
    plt.figure(figsize=(30, 30))
    plt.imshow(mask_grid, cmap='viridis')

    # Add text annotations for input_ids
    for i in range(grid_size):
        for j in range(grid_size):
            idx = i * grid_size + j
            if idx < len(mask_array):
                # Only show text for valid positions (not padding)
                text = tokenizer.decode(ids_grid[i, j])
                plt.text(j, i, text, 
                        ha="center", va="center", 
                        color="white" if mask_grid[i, j] < 0.5 else "black",
                        fontsize=7)

    plt.colorbar(label='Attention Value')
    plt.title('Attention Mask with Input IDs')
    plt.grid(False)
    plt.tight_layout()
    plt.show()

    # # Display a smaller section around the transition point (if any)
    # transition_points = np.where(np.diff(mask_array) != 0)[0]
    # if len(transition_points) > 0:
    #     # Get the first transition point
    #     trans_point = transition_points[0]
    #     # Display a window of tokens around the transition
    #     window = 20  # tokens before and after transition
    #     start_idx = max(0, trans_point - window)
    #     end_idx = min(len(mask_array), trans_point + window)
        
    #     # Create a zoomed-in view
    #     plt.figure(figsize=(12, 6))
    #     plt.imshow(mask_array[start_idx:end_idx].reshape(1, -1), cmap='viridis', aspect='auto')
        
    #     # Add token IDs as text
    #     for i, idx in enumerate(range(start_idx, end_idx)):
    #         plt.text(i, 0, str(input_ids_array[idx]), 
    #                 ha="center", va="center", 
    #                 color="white" if mask_array[idx] < 0.5 else "black",
    #                 fontsize=9)
        
    #     plt.title(f'Zoomed View Around Transition Point (position {trans_point})')
    #     plt.yticks([])
    #     plt.xlabel('Token Position')
    #     plt.grid(False)
    #     plt.tight_layout()
    #     plt.show()
```
